Подход: можно использовать churn rate, чтобы оценить отток пользователей в 2016 и 2017 году. Также может быть полезен retentation rate, чтобы узнать, как много клиентов всё ещё остаётся в 2017 году и сравнить с показателями 2016 года

```python
import pandas as pd
pd.__version__
```

```python
orders = pd.read_csv(
    'Studies/orders.csv'
)
orders
```

```python
customers = pd.read_csv(
    'Studies/customers.csv'
)
customers
```

Предобработка данных: убираем дубликаты и пустые значения 

```python
orders.drop_duplicates(inplace=True)
customers.drop_duplicates(inplace=True)
```

```python
orders=orders[(orders.customer_id.notnull()) & (orders.sales!=0)]
```

```python
orders['order_date'] = pd.to_datetime(orders['order_date'], infer_datetime_format=True)
orders['order_date'] = orders['order_date'].dt.strftime('%Y-%m') 
```

Разедлим данные на 2016 год и 2017 и проанализируем кол-во пользователей 

```python
orders2016 = orders.loc[(orders['order_date'] >= '2016-01')
                     & (orders['order_date'] < '2017-01')]
orders2016
```

```python
orders2017 = orders.loc[(orders['order_date'] >= '2017-01')
                     & (orders['order_date'] <= '2017-12')]
orders2017
```

Здесь я смотрю, насколько в целом заказов в 2017 году меньше, чем в 2016 году

```python
count2016 = orders.loc[(orders['order_date'] >= '2016-01')
                     & (orders['order_date'] <= '2016-12')]['id'].count()
count2017 = orders.loc[(orders['order_date'] >= '2017-01')
                     & (orders['order_date'] <= '2017-12')]['id'].count()
churn_rate_orders = count2017 - count2016
churn_rate_orders
```

Создадим таблицы с количеством клиентов по месяцам в 2016 и в 2017 году

```python
orders2016_cnt = orders2016.groupby(['order_date'], as_index=False )['customer_id'].nunique()
orders2016_cnt
```

```python
orders2017_cnt = orders2017.groupby(['order_date'], as_index=False )['customer_id'].nunique()
orders2017_cnt
```

```python
prev_cnt = 157
for index, row in orders2017_cnt.iterrows():
    cnt = (prev_cnt - row['customer_id'])/prev_cnt
    print(f'Churn rate at month {index + 1}: {cnt:.2f}')
    prev_cnt = row['customer_id']
```

```python
prev_cnt = 46
for index, row in orders2016_cnt.iterrows():
    cnt = (prev_cnt - row['customer_id'])/prev_cnt
    print(f'Churn rate at month {index + 1}: {cnt:.2f}')
    prev_cnt = row['customer_id']
```

Теперь можно посмотреть на churn_rate в 2016 и в 2017 году по месяцам. Как видно из данных, большое кол-во клиентов ушло в период с 2016-12 по 2017-01 (было 176 клиентов в конце 2016, стало 69 в начале 17 года). Однако по churn rate видно, что последующие месяцы 2017 года в целом были неплохие с точки зрения оттока клиентов (в 6 месяцах мы видим отрицательное значение churn rate, значит, кол-во клиентов даже увеличивалось, причем в 11 месяцу -> 0.59 по сравнению с 10 месяцем; на 3 месяце -> 1.17 по сравнению со 2-м). Я бы выдвинула следующую гипотезу: проблема с доставкой всё-таки повлияла на отток пользователей в начале года, однако на последующие месяцы она влияния, вероятно, не оказывала (или улучшился какой-то другой фактор/ люди свыклись с такой доставкой). Кроме того, максимальное кол-во клиентов в 2017 году больше максимального кол-ва клиента в 2016. Видно, что 9 и 11 месяцы хорошие с точки зрения притока клиентов, и значения в 2017 году даже лучше значений 2016-го года.
